```python
# imports

import numpy as np
import matplotlib
import matplotlib.pyplot as plt



from vdw_evolve import allign_along_10

matplotlib.use('nbagg')
```

```python
# useful functions

def extract_molecule(xyz_path):
    molec = open(xyz_path).read()
    molec = molec.split()
    nr_atoms = int(molec[0])
    atoms = {}
    
    for atom_n in range(0,nr_atoms):
        atom_n = atom_n*4+1
        atom = molec[atom_n]
        atoms[atom] = {}
        atoms[atom]["x"] = float(molec[atom_n+1])
        atoms[atom]["y"] = float(molec[atom_n+2])
        atoms[atom]["z"] = float(molec[atom_n+3])
    return atoms

def t_units_to_x(molec, cel):
    a1 = np.array([cel[0,0],cel[1,0]])
    a2 = np.array([cel[1,0],cel[1,1]])
    new_m ={}
    for atom in molec:
        print(a1*molec[atom]["x"])
        r = molec[atom]["x"]*a1 + molec[atom]["y"]*a2
        new_m[atom]=r.T
    return new_m

def a1a2(cel):
    a1 = np.array([cel[0,0],cel[1,0]])
    a2 = np.array([cel[0,1],cel[1,1]])
    return a1,a2

def atoms_xy(molec):
    x=[]
    y=[]
    l=[]
    for atom in molec:
        x.append(molec[atom][0])
        y.append(molec[atom][1])
        l.append(atom)
    return x,y, l
```

```python
def get_scales(cel1,cel0):
    
    c0 = np.array([0,0])
    c1 = np.array([cel1[0][0],cel1[1][0]])
    c2 = np.array([cel1[0][1],cel1[1][1]])
    c3 = c1+c2
    
    c_ = [c0 ,c1,c2,c3]
    x = [c0[0], c1[0], c2[0], c3[0]]
    y = [c0[1], c1[1], c2[1], c3[1]]
    
    d = [np.sqrt(c[0]**2+c[1]**2) for c in c_]
    

    a1s = int(max(d)/np.sqrt(cel0.T[0][0]**2+cel0.T[0][1]))+1
    a2s = int(max(d)/np.sqrt(cel0.T[1][0]**2+cel0.T[1][1]))+1
    
    return a1s, a2s
```

```python
def plt_cel(cel, c,label=""):
    c0 = np.array([0,0])
    c1 = np.array([cel[0][0],cel[1][0]])
    c2 = np.array([cel[0][1],cel[1][1]])
    c3 = c1+c2
    plt.plot([c0[0], c1[0]], [c0[1], c1[1]], color=c,label=label, alpha =0.2)
    plt.plot([c0[0], c2[0]], [c0[1], c2[1]], color=c, alpha =0.2)
    plt.plot([c1[0], c3[0]], [c1[1], c3[1]], color=c, alpha =0.2)
    plt.plot([c2[0], c3[0]], [c2[1], c3[1]], color=c, alpha =0.2)
    

def plt_grid(cel0,a1s,a2s,label, c):
    
    
    plt.plot([-a1s*cel0[0][0], a1s*cel0[0][0]], [-a1s*cel0[1][0], a1s*cel0[1][0]], color=c,label=label, alpha =0.2)
    plt.plot([-a2s*cel0[0][1], a2s*cel0[0][1]], [-a2s*cel0[1][1], a2s*cel0[1][1]],  color=c, alpha =0.2)
    
    for k in range(-a1s,a1s):
        plt.plot([-a2s*cel0[0][1]+k*cel0[0][0], a2s*cel0[0][1]+k*cel0[0][0]], 
                 [-a2s*cel0[1][1]+k*cel0[1][0], a2s*cel0[1][1]+k*cel0[1][0]],  
                  color=c, 
                  alpha =0.2)
    for k in range(-a2s,a2s):
        plt.plot([-a1s*cel0[0][0]+k*cel0[0][1], a1s*cel0[0][0]+k*cel0[0][1]], 
                 [-a1s*cel0[1][0]+k*cel0[1][1], a1s*cel0[1][0]+k*cel0[1][1]], 
                 color=c, alpha =0.2)
        

def atom_grid(molec, cel0,a1s,a2s):
    atoms = {}
    
    for e,atom in enumerate(molec):
        atom_x = molec[atom][0]
        atom_y = molec[atom][1]
        counter= 1
        for i in range(-a1s,a1s):
             for j in range(-a2s,a2s):
                x=atom_x+i*cel0.T[0][0]+j*cel0.T[1][0]
                y=atom_y+i*cel0.T[0][1]+j*cel0.T[1][1]
                atoms[str(counter)+"_"+atom]=[x,y]
                counter+= 1
    return atoms
                
def plt_atom_grid(molec, cel0,a1s,a2s,label, c):
    
    for e,atom in enumerate(molec):
        atom_x = molec[atom][0]
        atom_y = molec[atom][1]
        x= []
        y= []
        for i in range(-a1s,a1s):
             for j in range(-a2s,a2s):
                x.append(atom_x+i*cel0.T[0][0]+j*cel0.T[1][0])
                y.append(atom_y+i*cel0.T[0][1]+j*cel0.T[1][1])
        plt.scatter(x,y, color= c[e],alpha = 0.2)
```

```python

a1s,a2s = get_scales(cel1,cel0)
origin = np.array([[0,0], [0,0]])

plt_grid(cel0,a1s,a2s,label="graphene", c='b')
plt.arrow(0, 0, cel0[0][0], cel0[1][0], head_width=0.05, head_length=0.5, color='b',label="a1_graphene", alpha =0.3)
plt.arrow(0, 0, cel0[0][1], cel0[1][1], head_width=0.05, head_length=0.5, color='b',label="a2_graphene", alpha =0.3)


plt_cel(cel1, c='r',label="super_cel")
plt.arrow(0, 0, cel1[0][0], cel1[1][0], head_width=0.5, head_length=0.5, color='r',label="a1s_cel", alpha =0.3)
plt.arrow(0, 0, cel1[0][1], cel1[1][1], head_width=0.5, head_length=0.5, color='r',label="a2s_cel", alpha =0.3)


plt_atom_grid(graphene0x, cel0,a1s,a2s,label="graphene", c=['b','r'])

# plt.legend()
plt.show()
```

```python
# v = t1C1+t2C2
# v0 = t1c10 + t2c20
# v1 = t1c11 + t2c21
# v0*c11 = t1c10c11 +t2c20c11
# v1*c10 = t1c11c10 +t2c21c10
# v0*c11-v1*c10= t2(c20c11-t2c21c10) => t2= (v0*c11-v1*c10)/(c20c11-c21c10)

def in_cell_check(v, cell):
    c1= cell.T[0]
    c2= cell.T[1]
    t2 = (v[0]*c1[1]-v[1]*c1[0])/(c2[0]*c1[1]-c2[1]*c1[0])
    t1= (v[0]-t2*c2[0])/c1[0]
    
    if 0<t1<=1 and 0<t2<=1:
        return True
    else:
        return False
```

```python
atomic_grid = atom_grid(graphene0x, cel0,a1s,a2s)

cell_atoms ={}
for atom in atomic_grid:
    if in_cell_check(atomic_grid[atom], cel1):
        cell_atoms[atom]=atomic_grid[atom]
        
```
